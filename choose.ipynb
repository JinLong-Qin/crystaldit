```python
import os
import random
import shutil
```

```python
def copy_random_cif_files(source_dir, target_dir, num_files=20):
    
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
        print(f"create target dir.: {target_dir}")
    
    cif_files = [f for f in os.listdir(source_dir) if f.endswith('.cif')]
    
    
    selected_files = random.sample(cif_files, num_files)

    count = 0
    for filename in selected_files:
        src_path = os.path.join(source_dir, filename)
        dst_path = os.path.join(target_dir, filename)
        shutil.copy2(src_path, dst_path)
        count += 1
    
```

```python
source_root_directory = "/ssd1/CrystalDiT/mattpuscibert_processed_text3_crystals"  
target_root_directory = "/ssd1/CrystalDiT/mattpuscibert_processed_text3_crystals_500"

# copy_random_cif_files(source_directory, target_directory)
```

```python
for subdir in os.listdir(source_root_directory):
    source_directory = os.path.join(source_root_directory, subdir)
    if len(os.listdir(source_directory))< 500:
        print(f'{source_directory} has {len(os.listdir(source_directory))} files.')
    target_directory = os.path.join(target_root_directory, subdir)
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)
        print(f"create target dir.: {target_directory}")
    if os.path.isdir(source_directory):
        copy_random_cif_files(source_directory, target_directory, num_files=500)
        print(f'Finished randomly copying 500 cif files from {source_directory} to {target_directory}.')
```
